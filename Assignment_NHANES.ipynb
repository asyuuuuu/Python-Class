```python
#data normalization

data1['ALQ130'] = (data1['ALQ130']-data1['ALQ130'].mean())/data1['ALQ130'].std()
data1['RIDAGEYR'] = (data1['RIDAGEYR']-data1['RIDAGEYR'].mean())/data1['RIDAGEYR'].std()
data1['DMDMARTL'] = (data1['DMDMARTL']-data1['DMDMARTL'].mean())/data1['DMDMARTL'].std()
data1['WTINT2YR'] = (data1['WTINT2YR']-data1['WTINT2YR'].mean())/data1['WTINT2YR'].std()
data1['SDMVSTRA'] = (data1['SDMVSTRA']-data1['SDMVSTRA'].mean())/data1['SDMVSTRA'].std()
data1['BPXSY1'] = (data1['BPXSY1']-data1['BPXSY1'].mean())/data1['BPXSY1'].std()
data1['BPXDI1'] = (data1['BPXDI1']-data1['BPXDI1'].mean())/data1['BPXDI1'].std()
data1['BPXSY2'] = (data1['BPXSY2']-data1['BPXSY2'].mean())/data1['BPXSY2'].std()
data1['BPXDI2'] = (data1['BPXDI2']-data1['BPXDI2'].mean())/data1['BPXDI2'].std()
data1['BMXWT'] = (data1['BMXWT']-data1['BMXWT'].mean())/data1['BMXWT'].std()
data1['BMXBMI'] = (data1['BMXBMI']-data1['BMXBMI'].mean())/data1['BMXBMI'].std()
data1['BMXLEG'] = (data1['BMXLEG']-data1['BMXLEG'].mean())/data1['BMXLEG'].std()
data1['BMXARML'] = (data1['BMXARML']-data1['BMXARML'].mean())/data1['BMXARML'].std()
data1['BMXARMC'] = (data1['BMXARMC']-data1['BMXARMC'].mean())/data1['BMXARMC'].std()
data1['BMXWAIST'] = (data1['BMXWAIST']-data1['BMXWAIST'].mean())/data1['BMXWAIST'].std()

data1[['ALQ130', 'RIDAGEYR', 'DMDMARTL']].head()
```
