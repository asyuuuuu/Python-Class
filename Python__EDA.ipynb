#Correlation matrix

```python
#correlation between numerical columns (relationship analysis)

df_numeric = df.select_dtypes(include = ['number'])
df_numeric.corr()

#understand r/ship between numerical variables
#measures how changes in one variable are associated with changes in another.

#positive correlation: as one variable increases, the other variable also increases.
#negative correlation: as one variable increases, the other variable decreases.
#no correlation: there is no relationship between the two variables.

#correlation coefficient: a value between -1 and 1 that measures the strength and direction of
```

```python
#selecting certain numerical features to compare with price

df[['engine-size', 'city-mpg', 'bore', 'stroke', 'horsepower', 'price']].corr()
```

```python
#positive correlation

from matplotlib import pyplot as plt

import seaborn as sns

sns.regplot(x ='engine-size', y = 'price', data = df)
```

```python
#negative correlation

sns.regplot(x = 'city-mpg', y = 'price', data = df)
```

```python
#weak correlation

sns.regplot(x ='stroke', y = 'price', data = df)
```

```python
#importing scipy to find out correlation and p value
import scipy.stats as st

correlation, p_value = st.pearsonr(df['engine-size'], df['price'])
print ('The correlation is:', correlation, 'and the p_value is:', p_value)
```

```python
correlation, p_value = st.pearsonr(df['city-mpg'], df['price'])
print('The correlation is:', correlation, ' and the p_value is:', p_value)
```

```python
correlation, p_value = st.pearsonr(df['stroke'], df['price'])
print('The correlation is:', correlation, ' and the p_value is:', p_value)
```

the data for stroke column not clean. contain missing value. for this step, the array must not contain any missing value. it must be clean.

```python
#thus, check the missing value in that column

df['stroke'].isnull().sum()
```

```python
#replace the missing value with the mean valus

import numpy as np

df['stroke'].replace(np.nan, df['stroke'].mean(), inplace = True)
```

```python
#run the  correlation function again

correlation, p_value = st.pearsonr(df['stroke'], df['price'])
print('The correlation is:', correlation, ' and the p_value is:', p_value)
```

#Group by

```python
#Groupby - compare more than 1 categorical feature with 1 numerical feature
#as_index = False (to be more understandable - since the data will be by index)

df1 = df[['drive-wheels', 'body-style', 'price']]
df1_group = df1.groupby(['drive-wheels', 'body-style'], as_index = False).mean()

df1_group
```

```python
#index = rows

df1_pivot = df1_group.pivot(index = 'drive-wheels', columns = 'body-style')

df1_pivot
```

```python
df1_pivot = df1_group.pivot (index = 'body-style', columns = 'drive-wheels')

df1_pivot
```

```python
#create a heatmap
sns.heatmap(df1_pivot, annot = True, cmap = 'RdBu')
```

```python
# to view the whole category

plt.figure(figsize = (18,12)) # to minimize the size - parameters (width, height)

sns.heatmap(df_numeric.corr(), annot = True, cmap = 'RdBu') # can set other colors - refer cheatsheet.
```

```python
df2 = df[['num-of-doors', 'aspiration', 'price']]
df2_group = df2.groupby(['num-of-doors', 'aspiration'], as_index = False).mean()
df2_pivot = df2_group.pivot(index = 'num-of-doors', columns = 'aspiration')
df2_pivot
```

```python
sns.heatmap(df2_pivot, annot = True, cmap = 'RdBu')
```
