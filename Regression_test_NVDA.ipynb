```python
import numpy as np

msk = np.random.rand(len(df)) < 0.8 # length of the dataset lower than 80% will be chosen randomly for testing

train = df[msk] #80% of the data

test = df[~msk]
```

```python
plt.scatter(train.Open, train.Return_Percentage, color = 'blue')
plt.scatter(test.Open, test.Return_Percentage, color = 'red') #difference between train n test

plt.xlabel('Open')
plt.ylabel('Return_Percentage')
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()

train_x = np.asanyarray(train[['Open']])   #convert engine to array. to variable train x
train_y = np.asanyarray(train[['Return_Percentage']])

#training

regr.fit(train_x, train_y)
```

```python
print ('The value of the slope is: ',  regr.coef_)
print ('The value of the intercept is: ', regr.intercept_)
```

```python
plt.scatter(train.Open, train.Return_Percentage,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r') #-r is for the colour of the line, can be change to -g, you will see green
plt.xlabel("Open")
plt.ylabel("Return")
```

```python
x = 400

y = regr.intercept_  + regr.coef_ * x

y
```

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['Open']])
test_y = np.asanyarray(test[['Return_Percentage']])

test_y_ = regr.predict(test_x) #prediction based on test_x

print ('The R2 is: ', round (((r2_score(test_y, test_y_))*100),2))  #the accuracy is 77%
print ('The MSE is: ', np.mean((test_y_ - test_y)**2)) #the error
```

```python
# Open vs Return


regr = linear_model.LinearRegression()

train_x = np.asanyarray(train[['Open']])
train_y = np.asanyarray(train[['Return_Percentage']])

#training
regr.fit(train_x, train_y)

#prediction
test_x = np.asanyarray(test[['Open']])
test_y = np.asanyarray(test[['Return_Percentage']])

test_y_ = regr.predict(test_x)

#evaluation
print ('The R2 is: ', round (((r2_score(test_y, test_y_))*100),2))
print ('The MSE is: ', round ((((np.mean((test_y_ - test_y)**2)))*100), 2))
print ('The MAE is: ', round ((((np.mean(np.absolute(test_y_ - test_y))))*100),2))
```

```python
#what can i coclude from this analysis

print("Based on the analysis, we can conclude that there's a weak positive correlation between the opening price ('Open') and the return percentage ('Return_Percentage') of NVDA stock. The linear regression model built on the training data explains approximately 77% of the variance in the return percentage. However, the mean squared error indicates that there's still a considerable amount of unexplained variance, suggesting that other factors beyond the opening price significantly influence the return percentage.")
```
